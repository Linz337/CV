# Opencv

1.导入库 import library

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

2.读取图像 Read image

```python
image = cv2.imread("image.jpg")
```

注：这里得到的是一个三维矩阵，高度height，宽度width，通道数channel（RGB是三通道）
cv2读取得到的是BRG图像，skimage读取得到的是RGB图像
！所以需要对BGR图像进行转换：

Note: What you get here is a three-dimensional matrix, height, width, channel number (RGB is three channels). What cv2 reads is a BRG image, and what skimage reads is an RGB image! So the BGR image needs to be converted:

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

再用
And then

```python
plt.imshow(image) #将图像显示在当前的图像窗口中 Display the image in the current image window
plt.show() #显示图片 Display pictures

# imshow 函数用于在 Matplotlib 的图像窗口中显示图像。
# show 函数用于显示 Matplotlib 图像窗口本身，它会将内存中渲染好的图像显示在屏幕上。
# 两者缺一不可

# The imshow function is used to display images in the image window of Matplotlib. 
# The show function is used to display the Matplotlib image window itself, which displays the rendered image in memory on the screen. 
# Both are indispensable
```
